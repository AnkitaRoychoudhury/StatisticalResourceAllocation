## Figure 3B: PLS-derived tradeoff axes for E. coli mRNA data

### Plot tradeoff axes

**Load growth data**

```python
df_hwa_gr = pd.read_excel('../data/e.coli/science.abk2066_table_s3.xlsx', sheet_name = 0)
growth_map = df_hwa_gr.set_index("Sample ID")["Growth rate (1/h)"]
```

**Load mRNA data**

```python
df_hwa_mrna = pd.read_excel("../data/e.coli/science.abk2066_table_s3.xlsx", sheet_name = 1)
df_hwa_mrna.head(2)
```

**Process mRNA df and add growth values**

```python
df_hwa_mrna = df_hwa_mrna.drop(['locus', 'gene length (nt)'], axis = 1)
df_hwa_mrna_T = df_hwa_mrna.T
df_hwa_mrna_T.columns = df_hwa_mrna_T.iloc[0].values

# drop sample with no growth rate
df_hwa_mrna_T = df_hwa_mrna_T.drop(['gene', 'a4_1.1']) 

# add growth rate values
df_hwa_mrna_T['growth_rate'] = df_hwa_mrna_T.index.map(growth_map)

df_hwa_mrna_T.head(2)
```

**Define separate dfs for each condition**

```python
# carbon
df_c = df_hwa_mrna_T[df_hwa_mrna_T.index.str.startswith('c')]
# ammonia
df_a = df_hwa_mrna_T[df_hwa_mrna_T.index.str.startswith('a')]
# chloramphenicol
df_r = df_hwa_mrna_T[df_hwa_mrna_T.index.str.startswith('r')]
```

**Plot**

```python
beta_df, p_scatter, p_hist = run_pls(df_c, t=0, condition = 'Carbon')
```

```python
beta_df, p_scatter, p_hist = run_pls(df_a, t=0, condition = 'Nitrogen')
```

```python
beta_df, p_scatter, p_hist = run_pls(df_r, t=0, condition = 'Chloramphenicol')
```

## Figure 3C: PLS-derived tradeoff axes for S. pombe mRNA data

### Plot tradeoff axes

**Load mRNA and growth data**

```python
df_kleijn_mrna = pd.read_excel('../data/s.pombe/normalized_mRNA_counts.xlsx')
df_kleijn_gr = pd.read_excel('../data/s.pombe/growth_info.xlsx')
```

**Process dfs**

```python
df_kleijn_mrna['sample rep'] = df_kleijn_mrna['medium'] + df_kleijn_mrna['replicate'].fillna(0).astype(int).astype(str)
df_kleijn_mrna = df_kleijn_mrna[['PomBaseID', 'sample rep', 'raw_counts']]

# remove last row, extraneous
df_kleijn_mrna = df_kleijn_mrna.iloc[:-1]

df_kleijn_gr['sample rep'] = df_kleijn_gr['medium'] + df_kleijn_gr['replicate'].astype(str)
```

**Merge dfs**

```python
df_kleijn_merge = df_kleijn_mrna.merge(df_kleijn_gr[['sample rep', 'growth_rate']], on='sample rep', how='left')
df_kleijn_merge = df_kleijn_merge.iloc[:-1]

# Pivot to wide format: rows = sample rep, columns = PomBaseID, values = raw_counts
df_wide = df_kleijn_merge.pivot_table(
    index='sample rep',
    columns='PomBaseID',
    values='raw_counts'
).reset_index()

# Get unique growth rate per sample rep
growth = df_kleijn_gr[['sample rep', 'growth_rate']].drop_duplicates()

# Merge growth_rate back into the wide-format DataFrame
df_wide = df_wide.merge(growth, on='sample rep', how='left')
df_wide.head(2)
```

**Plot PLS with no thresholding**

```python
df_input = df_wide.set_index('sample rep')
beta_df, p_scatter, p_hist = run_pls(df_input, t=0, condition='s.pombe')
```

**Run permutation test to identify threshold**

```python
# define data
y = df_input['growth_rate'].values
X = df_input.drop(columns=['growth_rate']).values
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# run permutation test
n_perms = 1000
null_betas = []

for _ in tqdm(range(n_perms)):
    y_perm = np.random.permutation(y)
    pls_perm = PLSRegression(n_components=2)
    pls_perm.fit(X_scaled, y_perm)
    null_betas.append(pls_perm.coef_.flatten())

null_betas = np.array(null_betas)  # shape (n_perms, p)

# threshold = 95th percentile of null |beta| distribution
tau = np.percentile(np.abs(null_betas), 95)
tau
```

**Plot PLS with thresholding**

```python
# use this threshold for consistency
# every permutation run gives a slightly different value

tau = 6.010127746160799e-05
```

```python
beta_df, p_scatter, p_hist = run_pls(df_input, t=6.010127746160799e-05, condition='S.pombe')
```
