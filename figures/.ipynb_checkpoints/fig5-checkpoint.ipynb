## Figure 5: Distinct regulatory regimes

```python
import numpy as np
import sys
sys.path.insert(0, '../analysis')

import bokeh.io
import bokeh.plotting
bokeh.io.output_notebook()

from model2 import (
    DEFAULT_ECOLI, DEFAULT_YEAST, DEFAULT_SHARED,
    solve_steady_state,
    plot_total_mrna, plot_ribosome_fraction, plot_elongation_rate,
)
```

## Figure 5C: Model outputs

**Define parameters**

```python
# Co-limitation regime (E. coli) ──────────────────────────────────────
ecoli_params = dict(
    alpha = 95,    # max translation rate
    beta  = 2,     # amino-acid production per metabolic protein
    eta   = 1,     # AA cost per ribosome cycle
    kappa = 10,    # metabolic capacity scalar
    delta = 130,   # tRNA decay rate
)

# ── Ribosomal mRNA limitation regime (S. cerevisiae) ────────────────────
yeast_params = dict(
    alpha = 2,
    beta  = 3000,
    eta   = 0.3,
    kappa = 0.01,
    delta = 130,
)

# ── Shared transcription parameters ──────────────────────────────────────
shared_params = dict(
    alpha_c    = 0.05,
    alpha_r    = 0.9,
    alpha_pol2 = 0.05,
    mq         = 1.0,
)

# ── Growth rate ranges ───────────────────────────────────────────────────
ecoli_growth     = np.linspace(0.002, 0.9, 70)
yeast_growth     = np.linspace(0.02,  0.7, 40)
yeast_growth_min = 0.1   # discard unphysical low-growth solutions
```

**Solve steady states**

```python
ecoli_data = solve_steady_state(ecoli_params, shared_params, ecoli_growth)
yeast_data = solve_steady_state(yeast_params, shared_params, yeast_growth,
                                growth_min=yeast_growth_min)
```

**Total mRNA fold change**

```python
p1, p2 = plot_total_mrna(ecoli_data, yeast_data)
bokeh.io.show(p1)
bokeh.io.show(p2)
```

**Ribosome fraction**

```python
p3, p4 = plot_ribosome_fraction(ecoli_data, yeast_data)
bokeh.io.show(p3)
bokeh.io.show(p4)
```

**Elongation rate**

```python
p5, p6 = plot_elongation_rate(ecoli_data, yeast_data)
bokeh.io.show(p5)
bokeh.io.show(p6)
```
