# Figure S2: Selection of PLS components via cross-validation

```python
import numpy as np
import pandas as pd

import bokeh.io
import bokeh.plotting
bokeh.io.output_notebook()
from bokeh.layouts import gridplot

import sys
sys.path.insert(0, '../analysis')
from plot_PLS_error import plot_errors_n
```

## Figure S2A: Model error across components for S. pombe

**Load and process data**

See `fig2.ipynb` for details on how files were generated

```python
df_bs_1 = pd.read_csv("../analysis/s.pombe_PLS_permutation_results/df_bs_n=1_500reps_seed=21.csv")
df_bs_2 = pd.read_csv("../analysis/s.pombe_PLS_permutation_results/df_bs_n=2_500reps_seed=21.csv")
df_bs_3 = pd.read_csv("../analysis/s.pombe_PLS_permutation_results/df_bs_n=3_500reps_seed=21.csv")
df_bs_4 = pd.read_csv("../analysis/s.pombe_PLS_permutation_results/df_bs_n=4_500reps_seed=21.csv")
df_bs_8 = pd.read_csv("../analysis/s.pombe_PLS_permutation_results/df_bs_n=8_500reps_seed=21.csv")

df_bs_1['identity'] = ['bs, n = 1'] * len(df_bs_1)
df_bs_2['identity'] = ['bs, n = 2'] * len(df_bs_2)
df_bs_3['identity'] = ['bs, n = 3'] * len(df_bs_3)
df_bs_4['identity'] = ['bs, n = 4'] * len(df_bs_4)
df_bs_8['identity'] = ['bs, n = 8'] * len(df_bs_8)


df_null_1 = pd.read_csv('../analysis/s.pombe_PLS_permutation_results/df_null_n=1_500reps_seed=21.csv', index_col = 0)
df_null_2 = pd.read_csv('../analysis/s.pombe_PLS_permutation_results/df_null_n=2_500reps_seed=21.csv', index_col = 0)
df_null_3 = pd.read_csv('../analysis/s.pombe_PLS_permutation_results/df_null_n=3_500reps_seed=21.csv', index_col = 0)
df_null_4 = pd.read_csv('../analysis/s.pombe_PLS_permutation_results/df_null_n=4_500reps_seed=21.csv', index_col = 0)
df_null_8 = pd.read_csv('../analysis/s.pombe_PLS_permutation_results/df_null_n=8_500reps_seed=21.csv', index_col = 0)


df_null_1['identity'] = ['null, n = 1'] * len(df_null_1)
df_null_2['identity'] = ['null, n = 2'] * len(df_null_2)
df_null_3['identity'] = ['null, n = 3'] * len(df_null_3)
df_null_4['identity'] = ['null, n = 4'] * len(df_null_4)
df_null_8['identity'] = ['null, n = 8'] * len(df_null_8)
```

```python
df_bs_merged = pd.concat([ df_bs_1, df_bs_2, df_bs_3, df_bs_4,df_bs_8], ignore_index = True)
df_null_merged = pd.concat([ df_null_1, df_null_2, df_null_3, df_null_4, df_null_8], ignore_index = True)

df_null_merged['num comps'] = df_null_merged['identity'].str.split().str[-1]
df_bs_merged['num comps'] = df_bs_merged['identity'].str.split().str[-1]
```

**Plot**

```python
seeds = [21]
#n_comps = [1, 2,3,8,12]
labels_short = ['mse', 'rmse', 'mae', 'r2']
plots = []

for s in seeds:
    for label in labels_short:
        
        plots.append(plot_errors_n(s, label, df_null_merged, df_bs_merged))
        #bokeh.io.show(plot_errors_n(s, label))
       # bokeh.io.show()


grid = gridplot([plots[:2], plots[2:4]], toolbar_location=None)
bokeh.io.show(grid)
```

## Figure S2B: Model error across components for E. coli

```python
df_bs_1 = pd.read_csv("../analysis/e.coli_PLS_permutation_results/df_bs_n=1_500reps_seed=21.csv")
df_bs_2 = pd.read_csv("../analysis/e.coli_PLS_permutation_results/df_bs_n=2_500reps_seed=21.csv")
df_bs_3 = pd.read_csv("../analysis/e.coli_PLS_permutation_results/df_bs_n=3_500reps_seed=21.csv")
df_bs_4 = pd.read_csv("../analysis/e.coli_PLS_permutation_results/df_bs_n=4_500reps_seed=21.csv")
df_bs_8 = pd.read_csv("../analysis/e.coli_PLS_permutation_results/df_bs_n=8_500reps_seed=21.csv")

df_bs_1['identity'] = ['bs, n = 1'] * len(df_bs_1)
df_bs_2['identity'] = ['bs, n = 2'] * len(df_bs_2)
df_bs_3['identity'] = ['bs, n = 3'] * len(df_bs_3)
df_bs_4['identity'] = ['bs, n = 4'] * len(df_bs_4)
df_bs_8['identity'] = ['bs, n = 8'] * len(df_bs_8)


df_null_1 = pd.read_csv('../analysis/e.coli_PLS_permutation_results/df_null_n=1_500reps_seed=21.csv', index_col = 0)
df_null_2 = pd.read_csv('../analysis/e.coli_PLS_permutation_results/df_null_n=2_500reps_seed=21.csv', index_col = 0)
df_null_3 = pd.read_csv('../analysis/e.coli_PLS_permutation_results/df_null_n=3_500reps_seed=21.csv', index_col = 0)
df_null_4 = pd.read_csv('../analysis/e.coli_PLS_permutation_results/df_null_n=4_500reps_seed=21.csv', index_col = 0)
df_null_8 = pd.read_csv('../analysis/e.coli_PLS_permutation_results/df_null_n=8_500reps_seed=21.csv', index_col = 0)


df_null_1['identity'] = ['null, n = 1'] * len(df_null_1)
df_null_2['identity'] = ['null, n = 2'] * len(df_null_2)
df_null_3['identity'] = ['null, n = 3'] * len(df_null_3)
df_null_4['identity'] = ['null, n = 4'] * len(df_null_4)
df_null_8['identity'] = ['null, n = 8'] * len(df_null_8)
```

```python
df_bs_merged = pd.concat([ df_bs_1, df_bs_2, df_bs_3, df_bs_4,df_bs_8], ignore_index = True)
df_null_merged = pd.concat([ df_null_1, df_null_2, df_null_3, df_null_4, df_null_8], ignore_index = True)

df_null_merged['num comps'] = df_null_merged['identity'].str.split().str[-1]
df_bs_merged['num comps'] = df_bs_merged['identity'].str.split().str[-1]
```

```python
seeds = [21]
labels_short = ['mse', 'rmse', 'mae', 'r2']
plots = []

for s in seeds:
    for label in labels_short:
        plots.append(plot_errors_n(s, label, df_null_merged, df_bs_merged))

grid = gridplot([plots[:2], plots[2:4]], toolbar_location=None)
bokeh.io.show(grid)
```
